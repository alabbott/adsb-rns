```python
import proto

url = 'https://adsb.alabbott.dev/data/aircraft.json'
# Generic Chicago Location as default - Delilah's 41.93230° N, 87.65804° W
clat = 41.93
clon = -87.65
range = 25

aircraft_list = proto.fetch_aircraft(url)
print(f'Filtered: {len(aircraft_list)}')

filtered = proto.filter_aircraft(aircraft_list, clat, clon, range)

print(f'Filtered: {len(filtered)}')

for ac in filtered:
    callsign = ac.get('flight') if ac.get('flight') is not None else ac.get('r')
    print(callsign)
```
